```python
def retrieve_cimis_station_info(verbose=False):
    StationNbr = []
    Name = []
    station_url = 'http://et.water.ca.gov/api/station'
    try:
        content = requests.get ('station_url')

        stations = content['Stations']
        for i in stations:
            if i['IsActive'] == "True":
                StationNbr.append(i['StationNbr'])
                Name.append(i['Name'])
        if verbose is True:
            return stations
        else:
            return dict(zip(StationNbr, Name))
    except:
        print ("HTTPError, Station info not available")
```

```python
def get_cimis_data(start_date, end_date, targets, app_key='f8eb6841-dace-48cb-b677-8ed1092a843f'):
    url =f'http://et.water.ca.gov/api/data?appKey={app_key}&targets={targets}&startDate={start_date}&endDate={end_date}&dataItems=hly-eto&unitOfMeasure=M'
    response =requests.get(url)
    if response.status_code == 200:
        data = response.json()
        if len(data['Data']['Providers']) == 1 and data['Data']['Providers'][0]['Name']  == 'cimis':
            records = data['Data']['Providers'][0]['Records']   
        elif len(data['Data']['Providers']) > 1:
            found_cimis = False
            provider_names = []
            for provider in data['Data']['Providers']:
                name = provider['Name']
                provider_names.append(name)
                if name == 'cimis':
                    records = ['Data']['Providers'][0]['Records']
                    found_cimis = True
            if not found_cimis:
                raise Exception (f"Found multiple data providers from API, none were 'cimis': {provider_names}")
                                                                                
        else:
            raise Exception("Problem finding cimis provider from API response")
        df = pd.DataFrame(records)
        for key in df['HlyEto'].iloc[0].keys():
            df[f'HlyETo_{key}'] = df['HlyEto'].apply(lambda x: x[key])
        return df
    else:
        raise Exception(f"problem contacting API: {print(response.status_code, response.content)}")
```
